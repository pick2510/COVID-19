```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import itertools


plt.rcParams["figure.figsize"] = [16,9]

from labellines import labelLine, labelLines
```

```python
data = pd.read_csv("../csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv")
```

```python
data.drop(["Province/State", "Lat", "Long"], axis =1, inplace=True)
```

```python
columns = list(data.columns)
columns[0] = "Country"
data.columns = columns
```

```python
data_grouped = data.groupby("Country").sum().T
data_grouped.columns
```

```python
ch_data = pd.read_csv("../covid19-cases-switzerland/covid_19_cases_switzerland_standard_format.csv").drop(
["lat", "long", "name_canton", "country"],axis=1)
ch_data
```

```python
cases = ch_data[["date","abbreviation_canton","total_currently_positive_cases"]].copy()
cases.date = pd.to_datetime(cases.date)
```

```python
grouped = cases.groupby("abbreviation_canton")
```

```python
for group in grouped:
    data = pd.DataFrame(group[1]).drop("abbreviation_canton", axis=1)
    data.index = data.date
    data.drop("date", axis=1, inplace=True)
    data = data.interpolate(method='linear')
    data = data.loc["2020-03-17":]
    data = data.dropna()
    data = data.loc[~(data==0).all(axis=1)]
    print(data, group[0])
    #print(data)
    plt.plot(data, marker=next(marker), label=group[0])
plt.yscale("log")
plt.legend()
```
