```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import itertools


plt.rcParams["figure.figsize"] = [16,9]

from labellines import labelLine, labelLines
```

```python
data = pd.read_csv("../csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv")
```

```python
data.drop(["Province/State", "Lat", "Long"], axis =1, inplace=True)
```

```python
columns = list(data.columns)
columns[0] = "Country"
data.columns = columns
```

```python
data_grouped = data.groupby("Country").sum().T
data_grouped.columns
```

```python
data_grouped[["France", "Japan", "Switzerland", "United Kingdom", "Germany"]].plot()
plt.grid()
```

```python
#re.findall("Ir",list(data_grouped.columns))
list(data_grouped.columns)
```

```python
ch = data_grouped[["Switzerland"]]
idx_ch = np.where(ch.Switzerland >= 100)

it = data_grouped[["Italy"]]
idx_it = np.where(it.Italy >= 100)

de = data_grouped[["Germany"]]
idx_de = np.where(de.Germany >= 100)

ir = data_grouped[["Iran"]]
idx_ir = np.where(ir.Iran >= 100)


ca = data_grouped[["Canada"]]
idx_ca = np.where(ca.Canada >= 100)

be = data_grouped[["Belgium"]]
idx_be = np.where(be.Belgium >= 100)

at = data_grouped[["Austria"]]
idx_at = np.where(at.Austria >= 100)


ko = data_grouped[["Korea, South"]]
idx_ko = np.where(ko["Korea, South"] >= 100)

us = data_grouped[["US"]]
idx_us = np.where(us.US >= 100)

uk = data_grouped[["United Kingdom"]]
idx_uk = np.where(uk["United Kingdom"] >= 100)
```

```python
marker = itertools.cycle(("o", "v", "^", ">", "<","8", "s","p", "h", "D")) 
```

```python
plt.plot(it.Italy.values[idx_it], marker=next(marker),label="Italy")
plt.plot(ch.Switzerland.values[idx_ch],marker=next(marker), label="CH")
plt.plot(de.Germany.values[idx_de], marker=next(marker),  label="Germany")
plt.plot(ir.Iran.values[idx_ir], marker=next(marker), label="Iran")
plt.plot(ca.Canada.values[idx_ca],marker=next(marker), label="Canada")
plt.plot(us.US.values[idx_us],marker=next(marker), label="US")
plt.plot(uk["United Kingdom"].values[idx_uk], marker=next(marker), label="UK")
plt.plot(be.Belgium.values[idx_be],marker=next(marker), label="Belgium")
plt.plot(at.Austria.values[idx_at], marker=next(marker), label="Austria")
plt.plot(ko["Korea, South"].values[idx_ko], marker=next(marker), label="South Korea")
plt.legend()
plt.yscale("log")
#plt.xlim([0,  len(idx_ch[0])])
#plt.ylim([0,4000])
plt.grid(which="both")
#labelLines(plt.gca().get_lines(),xvals=(0.1, 10),zorder=2.5,align=False,fontsize=8)
plt.xlabel("Days since cases > 100")
plt.ylabel("Cases")
plt.title("Cases in important Chair of Building Physics countries from t = 0 where cases > 100")
plt.tight_layout()
plt.savefig("cbp_cases_log.png")
```
